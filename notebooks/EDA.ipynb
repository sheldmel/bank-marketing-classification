```python
# pdays
df = X.copy()
not_contacted = df[df['pdays'] == -1]['pdays']
df = df[df['pdays'] != -1]['pdays']

counts = df.value_counts().to_frame().reset_index()
contacted_count = counts.sum()['count']
not_contacted_count = not_contacted.value_counts()[-1]

print(f'Count of no previous contact: {not_contacted_count}')
print(f'Count of previous contact: {contacted_count}\n')

plt.figure(figsize=(7,4))
plt.hist(df, bins=70, edgecolor='black')
plt.xlabel('pdays')
plt.ylabel('Count')
plt.title('Distribution of pdays (excluding not previously contacted)')
plt.show()
```

```python
# poutcome
df = X.copy()['poutcome']
counts = df.value_counts()

sum = 0
for i in counts:
  sum += i

percent = counts['success']/sum*100
print(f'success percentage: {percent:.2f}%')

counts.plot(kind='bar')
plt.xlabel('poutcome')
plt.ylabel('Count')
plt.title('Distribution of poutcome')
plt.show()
```

```python
# y (has the client subscribed a term deposit?)
df = y.copy()['y']
counts = df.value_counts()

percent = counts['yes']/(counts['no']+counts['yes'])*100
print(f'success percentage: {percent:.2f}%\n')

counts.plot(kind='bar')
plt.xlabel('has the client subscribed a term deposit?')
plt.ylabel('Count')
plt.title('Distribution of y')
plt.show()
```

Correlations:

```python
df = X.copy()
df.apply(lambda x: x.factorize()[0]).corr()
```

```python


# Convert categorical features to numeric (factorize) for correlation
df = X.copy()
corr = df.apply(lambda x: x.factorize()[0]).corr()

# Turn correlation matrix into long form
corr_pairs = corr.abs().unstack().reset_index()
corr_pairs.columns = ['Feature 1', 'Feature 2', 'Absolute Correlation']

# Remove self-correlations and duplicates
corr_pairs = corr_pairs[corr_pairs['Feature 1'] != corr_pairs['Feature 2']]
corr_pairs = corr_pairs.drop_duplicates(subset='Absolute Correlation')

# Sort correlations
sorted_corr = corr_pairs.sort_values(by='Absolute Correlation', ascending=False)

# Top 10 and least 10
top10 = sorted_corr.head(10)
least10 = sorted_corr.tail(10).iloc[::-1]

# Display tables
print("Top 10 Most Correlated Feature Pairs:\n")
print(top10.to_string(index=False))

print("\nLeast 10 Correlated Feature Pairs:\n")
print(least10.to_string(index=False))
```
